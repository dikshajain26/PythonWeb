```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('Mamaearth.csv')

# Clean the "MRP" column by removing non-numeric characters and converting to float
df['MRP'] = df['MRP'].str.replace('Rs', '')  # Remove 'Rs'
df['MRP'] = df['MRP'].str.replace(',', '')     # Remove commas
df['MRP'] = df['MRP'].str.extract(r'(\d+\.\d+)').astype(float)  # Extract and convert to float

# Assuming 'Rating' is the column name containing ratings, convert it to numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Assuming 'Reviews' is the column name containing review counts, replace missing values with zeros
df['Reviews'].fillna(0, inplace=True)

# Assuming 'Reviews' is the column name containing review counts, convert it to numeric
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

# Get the total number of products (assuming 'Product' is the column name containing product names)
total_products = len(df['Product Name'])

# Calculate the average of the 'MRP' column
average_mrp = df['MRP'].mean()

# Calculate the average rating (ignoring non-numeric values)
average_rating = df['Rating'].mean()

# Calculate the average number of reviews (ignoring non-numeric values)
average_reviews = df['Reviews'].mean()

print(f'Total number of products: {total_products}')
print(f'Average MRP: {average_mrp:.2f}')
print(f'Average Rating: {average_rating:.2f}')
print(f'Average Reviews: {average_reviews:.2f}')
```

```python
import pandas as pd
import scipy.stats as stats

# Load the CSV file into a DataFrame (replace 'Mamaearth.csv' with your actual file name)
df = pd.read_csv('Mamaearth.csv')

# Define the product categories
categories = df['Category'].unique()

# Create an empty list to store the samples for each category
category_samples = []

# Split the dataset into samples by category
for category in categories:
    category_samples.append(df[df['Category'] == category]['Rating'])

# Perform one-way ANOVA test
f_statistic, p_value = stats.f_oneway(*category_samples)

# Set the significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha to determine significance
if p_value < alpha:
    print("The differences in Ratings between categories are statistically significant (reject null hypothesis).")
else:
    print("There is no significant difference in Ratings between categories (fail to reject null hypothesis).")

# Print the F-statistic and p-value
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")
```

```python
import pandas as pd
import scipy.stats as stats

# Load the CSV file into a DataFrame (replace 'Mamaearth.csv' with your actual file name)
df = pd.read_csv('Mamaearth.csv')

# Specify the two categories you want to compare
category1 = 'Skincare'
category2 = 'Haircare'

# Filter the DataFrame for the selected categories
category1_mrp = df[df['Category'] == category1]['MRP']
category2_mrp = df[df['Category'] == category2]['MRP']

# Perform a two-sample t-test to compare the average MRP between the two categories
t_statistic, p_value = stats.ttest_ind(category1_mrp, category2_mrp, equal_var=False)

# Set the significance level (alpha)
alpha = 0.05

# Check if the p-value is less than alpha to determine significance
if p_value < alpha:
    print(f"The average MRP for {category1} and {category2} are significantly different (reject null hypothesis).")
else:
    print(f"There is no significant difference in the average MRP between {category1} and {category2} (fail to reject null hypothesis).")

# Print the t-statistic and p-value
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")
```

```python
import pandas as pd
import statsmodels.api as sm

# Load the CSV file into a DataFrame (replace 'Mamaearth.csv' with your actual file name)
df = pd.read_csv('Mamaearth.csv')

# Clean and extract the numeric portion of the "MRP" column
df['MRP'] = df['MRP'].str.replace('Rs', '').str.replace(',', '').str.extract(r'(\d+\.\d+)').astype(float)

# Define the dependent variable (Y) and independent variable (X)
Y = df['Rating']
X = df['MRP']

# Add a constant (intercept) to the independent variable
X = sm.add_constant(X)

# Fit the linear regression model
model = sm.OLS(Y, X).fit()

# Print the regression summary
print(model.summary())
```

```python
import pandas as pd
import scipy.stats as stats

# Load the CSV file into a DataFrame (replace 'Mamaearth.csv' with your actual file name)
df = pd.read_csv('Mamaearth.csv')

# Create a contingency table (cross-tabulation) between 'Category' and 'Discount'
contingency_table = pd.crosstab(df['Category'], df['Discount'])

# Perform the chi-square test for independence
chi2, p, _, _ = stats.chi2_contingency(contingency_table)

# Set the significance level (alpha)
alpha = 0.05

# Print the chi-square statistic and p-value
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")

# Check if the p-value is less than alpha to determine independence
if p < alpha:
    print("There is a significant relationship between 'Category' and 'Discount' (reject null hypothesis).")
else:
    print("There is no significant relationship between 'Category' and 'Discount' (fail to reject null hypothesis).")
```
